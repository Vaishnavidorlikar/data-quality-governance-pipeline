## 5. Performance Analysis

```python
# Performance testing
import time

def measure_performance(func, *args, **kwargs):
    """Measure execution time of a function"""
    start_time = time.time()
    result = func(*args, **kwargs)
    end_time = time.time()
    
    return {
        'result': result,
        'execution_time': end_time - start_time
    }

# Test different operations
performance_tests = {
    'Email Processing': lambda: email_agent.process_email(
        test_email['content'], test_email['sender'], test_email['subject']
    ),
    'Report Generation': lambda: report_agent.generate_report(
        test_data, 'monthly', 'Test Report'
    ),
    'Text Summarization': lambda: summarizer.summarize_text(
        test_text, 'general'
    ),
    'LLM Response': lambda: llm_client.generate_response(
        "Test prompt for performance measurement"
    )
}

performance_results = {}

for test_name, test_func in performance_tests.items():
    print(f"Running {test_name}...")
    result = measure_performance(test_func)
    performance_results[test_name] = result['execution_time']
    print(f"  Execution time: {result['execution_time']:.3f} seconds")

print("\nPerformance Summary:")
for test_name, exec_time in performance_results.items():
    print(f"{test_name}: {exec_time:.3f}s")
```

```python
# Visualize performance results
if performance_results:
    plt.figure(figsize=(10, 6))
    
    tests = list(performance_results.keys())
    times = list(performance_results.values())
    
    plt.bar(tests, times, color='skyblue')
    plt.title('Performance Test Results')
    plt.xlabel('Operation')
    plt.ylabel('Execution Time (seconds)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("No performance data to visualize")
```
